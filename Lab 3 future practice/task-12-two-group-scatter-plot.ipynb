```python
import matplotlib.pyplot as plt

#data 1
country = ['Canada', 'Australia', 'Saudi-Arabia', 'Italy', 'Germany']
rate = [43, 76, 64, 32, 27]

#data 2 
category = ["Dhaka", "Cumilla", "Rajshahi","Dinajpur"]
data = [20, 25, 17, 19]
#scatter plot 1st one
plt.scatter(country, rate, color="red")
plt.xlabel('Country Name')
plt.ylabel('Rate')
plt.title('Visa Acceptance Rate')
plt.show()
#scatter plot 2nd one
plt.scatter(category, data, color='green')
plt.xlabel('City Name')
plt.ylabel('Rate')
plt.title('Crime Rate in BD')
plt.show()
```
