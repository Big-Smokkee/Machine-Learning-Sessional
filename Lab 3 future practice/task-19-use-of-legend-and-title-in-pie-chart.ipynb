```python
import matplotlib.pyplot as plt

labels = ['First', "Second", "Third", "Fourth", 'Fifth', 'Sixth']
grades = [3.87, 3.96, 3.98, 3.72, 3.88, 3.80]
explode = (0.1,0.2,0.3,0.0, 0.1, 0.0)
colors = ['green', 'gray','yellow', 'red','purple', 'lime']

plt.title("Task 19 pie chart")
plt.pie(grades, labels=labels, explode=explode, colors=colors, shadow=True)
plt.legend(labels, title = "Semester number")
plt.show()
```
