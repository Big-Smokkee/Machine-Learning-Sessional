```python
import numpy as np
b = np.array([1,2,30])
print("Second element of the array is: ", b[2])
```

```python
import numpy as np
b = np.array([[1,2,30], [5, 60, 9]])
print("Second element of the 2D array is: ", b[1][1])
```

```python
import numpy as np
b = np.array([[[1,2,30], [5, 60, 9]], [[70,8,0], [9, 75, 89]]])
print("Second element of the 3D array is: ", b[1][1][2])
```
