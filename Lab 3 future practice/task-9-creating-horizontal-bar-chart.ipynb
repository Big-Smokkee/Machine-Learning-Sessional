```python
import matplotlib.pyplot as plt

#data
country = ['Canada', 'Australia', 'Saudi-Arabia', 'Italy', 'Germany']
rate = [43, 76, 64, 32, 27]

#ploting into bar chart
plt.barh(country,rate, color='cyan')
plt.title("Visa Acceptance Rate")
plt.xlabel("Acceptance Rate")
plt.ylabel("Country Name")
plt.show()
```
